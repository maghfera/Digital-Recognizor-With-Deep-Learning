```python
import tensorflow as tf
from tensorflow import keras
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x = train.drop('label', axis = 1)
y = train['label']
```

```python
#splittig data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7)
```

# Deep Learning Model

```python
model = keras.models.Sequential([
 keras.layers.Flatten(input_shape=[28, 28]),
 keras.layers.Dense(300, activation="relu"),
 keras.layers.Dense(100, activation="relu"),
 keras.layers.Dense(10, activation="softmax")
])
```

```python
model.summary()
```

```python
model.layers
```

```python
#define first hidden layer 
hidden1 = model.layers[1]
```

```python
 #accesing of weights and biases of dense layers
weights, biases = hidden1.get_weights()
weights
```

```python
#bais matrix should be zeros
biases
```

# Compiling the model

```python
model.compile(loss="sparse_categorical_crossentropy",
 optimizer="sgd",
 metrics=["accuracy"])
```

```python
#keras.utils.to_categorical() 
```

```python
x_valid, x_train = x[:5000] / 255, x[5000:] / 255
y_valid, y_train = y[:5000], y[5000:]
```

```python
history = model.fit(x_train, y_train, epochs=30, validation_data=(x_valid, y_valid))
```

```python
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) 
plt.show()
```

Both training acc and validation acc are increasing, and the training loss and validation loss are decreasing, so that refer to the overfitting is very small 😃😃  

```python
model.evaluate(x_test, y_test)
```
